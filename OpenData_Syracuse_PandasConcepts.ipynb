```python
#Replace using mean, median and mode
df = pd.read_csv('CrimeSyr.csv')
#print(df.columns)
#we don't have any column with a value that can be filled using mean, median or mode so let's consider using Fid for the timebeing
print(df.FID)
#Understand the O/P : 0,1,2,3...are the DF indices, and 1,2,3,4..are the FIDs, now ... represent the skipped rows from the O/p. total rows are 5628.
```

```python
#let's delete a FID at index 3 to fill in later
# To delete the row at index 3
df = df.drop(index=3)

# If you want to reset the index after dropping the row
df = df.reset_index(drop=True)
print(df.FID)
```

```python
# Create a DataFrame with the missing rows
missing_data = pd.DataFrame({
    'FID': [4, 5, 6, 7],
    # Add other columns that were in your original dataset
}, index=[3, 4, 5, 6])  # Specify where to insert

''' Let's Learn a basic biscuit - the difference between list, tuple & dictionary
List- [] , cannot have key-value pair like ID:[1,23,4] but eg - mu_list = [12,4]
Tuple- () , cannot have key-value pair like ID:[1,23,4] but eg - mu_tuple = (12,4)
Dictinary - {}, can have key-value pair; like {'ID' : [2,3,4]}
''' #anyways long-story short, missing_data is creating a dictionary for DF

# Concatenate with your current DataFrame
df_restored = pd.concat([df.iloc[:3], missing_data, df.iloc[3:]]).reset_index(drop=True)
print(df_restored.FID)
```
